```python
%matplotlib inline
import torch
from d2l import torch as d2l
```

## ReLU和Sigmoid  
由于ReLU和Sigmoid激活函数的性质不同，导致它们在训练过程中对学习率的敏感度也不同。  
*使用ReLU的网络需要较小的学习速率，而使用Sigmoid的网络可以使用较大的学习速率*

**ReLU函数**  
 - ReLU函数的输出范围是[0, +∞)，其导数要么是0，要么是1（输入为正数时）。  
 - ReLU没有梯度消失问题，但会遇到“梯度爆炸”问题，即如果学习率过大，梯度可能会变得非常大，导致模型参数更新过度。  
- 为了防止梯度爆炸，通常需要使用较小的学习速率。  

```python
# Rectified linear unit, ReLU(修正线性单元)
x = torch.arange(-8.0, 9.0, 0.1, requires_grad=True)
y = torch.relu(x)
d2l.plot(x.detach(), y.detach(), 'x', 'relu(x)', figsize=(5, 2.5))
```

```python
y.backward(torch.ones_like(x), retain_graph=True)
d2l.plot(x.detach(), x.grad, 'x', 'grad of relu', figsize=(5, 2.5))
```

**Sigmoid函数**  
  - Sigmoid函数的输出范围是(0, 1)，其导数最大值是0.25（在输入为0时）。
  - 因此，梯度在传播过程中会逐渐变小，这被称为“梯度消失”问题。
  - 由于梯度较小，更新步长较小，因此可以使用较大的学习速率来加速训练。

```python
#sigmoid函数:sigmoid(x)=1/(1+exp(−x))
y=torch.sigmoid(x)
d2l.plot(x.detach(), y.detach(), 'x', 'sigmoid(x)', figsize=(5, 2.5))
```

```python
x.grad.data.zero_()
y.backward(torch.ones_like(x), retain_graph=True)
d2l.plot(x.detach(), x.grad, 'x', 'grad of sigmoid', figsize=(5, 2.5))
```

```python
y = torch.tanh(x)
d2l.plot(x.detach(), y.detach(), 'x', 'tanh(x)', figsize=(5, 2.5))
```

```python
x.grad.data.zero_()
y.backward(torch.ones_like(x), retain_graph=True)
d2l.plot(x.detach(), x.grad, 'x', 'grad of tanh', figsize=(5, 2.5))
```
